```python
import pandas as pd
```

```python
data = pd.read_csv('titanic_train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
uniqueSex = data['Sex'].unique()
```

```python
uniqueTicket = data['Ticket'].unique()
uniqueCabin = data['Cabin'].unique()
uniqueEmbarked = data['Embarked'].unique()
```

```python
print(len(uniqueTicket))
print(len(uniqueCabin))
print(len(uniqueEmbarked))
```

```python
uniqueSex = uniqueSex.tolist()
uniqueTicket = uniqueTicket.tolist()
uniqueCabin = uniqueCabin.tolist()
uniqueEmbarked = uniqueEmbarked.tolist()
```

```python
data.reset_index(inplace=True)
```

```python
for i in range(0, len(data)):
    data['Sex'][i] = uniqueSex.index(data['Sex'][i])
    data['Ticket'][i] = uniqueTicket.index(data['Ticket'][i])
    data['Cabin'][i] = uniqueCabin.index(data['Cabin'][i])
    data['Embarked'][i] = uniqueEmbarked.index(data['Embarked'][i])
```

```python
data.tail()
```

```python
Y = data['Survived']
X = data[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']]
```

```python
X
```

```python
X['Sex'].describe()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X, Y)
```

```python
print('Accuracy', model.score(X, Y) * 100, '%')
```
